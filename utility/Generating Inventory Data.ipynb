```python
bom_projection = pd.merge(_cart, recipe, how = "inner", on="food_item_id")
bom_projection["quantity_consumed"] = bom_projection["ingredient_quantity"]*bom_projection["qty"]
bom_projection = bom_projection[["order_placed_at","order_id","ingredient_id","quantity_consumed"]]
bom_projection.tail()
```

```python
bom_projection["date"] = bom_projection.apply(lambda x: x["order_placed_at"].date(), axis = 1)
bom_projection_agg = bom_projection.groupby(by = ["ingredient_id","date"])["quantity_consumed"].sum()
bom_projection_agg
```

```python
bom_projection_agg = bom_projection_agg.groupby(by = "ingredient_id").mean()
# bom_projection_agg.reset_index(inplace=True, drop = False)
bom_projection_agg = bom_projection_agg.round(1)
bom_projection_agg
```

```python
bom_projection_agg = pd.DataFrame(bom_projection_agg).reset_index(drop=False)
```

```python
import datetime

date_axis = bom_projection[["date","ingredient_id"]].drop_duplicates().reset_index(drop=True)
inflow = pd.merge(bom_projection_agg,date_axis,on="ingredient_id",how="inner")
inflow["order_placed_at"] = inflow.apply(lambda x: datetime.datetime(x["date"].year, 
                                                                     x["date"].month, 
                                                                     x["date"].day, 
                                                                     hour =12, 
                                                                     minute=0, 
                                                                     second=0),
                                         axis = 1)
inflow
```

```python
inflow = inflow[["order_placed_at","ingredient_id","quantity_consumed","date"]]
inflow.head()
```

```python
bom_projection["quantity_consumed"] = bom_projection["quantity_consumed"]*-1
inventory_flow = pd.concat([bom_projection[["order_placed_at","ingredient_id","quantity_consumed","date"]], inflow])
inventory_flow
```

```python
inventory_stockouts = inventory_flow.groupby(by = ["ingredient_id","date"])["quantity_consumed"].sum()
inventory_stockouts = inventory_stockouts[inventory_stockouts < 0]
inventory_stockouts = inventory_stockouts*-1
inventory_stockouts = pd.DataFrame(inventory_stockouts).reset_index(drop=False)
inventory_stockouts
```

```python
inventory_stockouts["order_placed_at"] = inventory_stockouts.apply(lambda x: datetime.datetime(x["date"].year, 
                                                                     x["date"].month, 
                                                                     x["date"].day, 
                                                                     hour =12, 
                                                                     minute=0, 
                                                                     second=0),
                                         axis = 1)
inventory_stockouts = inventory_stockouts[["order_placed_at","ingredient_id","quantity_consumed","date"]]
```

```python
inventory_stockouts
```

```python
inventory_transactions = pd.concat([inventory_flow, inventory_stockouts])
inventory_transactions.sort_values(by="order_placed_at",inplace= True)
inventory_transactions
```

```python
inventory_transactions = pd.DataFrame(inventory_transactions.groupby(by=["order_placed_at","ingredient_id"])["quantity_consumed"].sum()).reset_index(drop=False)
```

```python
inventory_transactions
```

```python
inventory = pd.DataFrame()
for group, data in inventory_transactions.groupby(by=['ingredient_id']):
    data['qty_available'] = data['quantity_consumed'].cumsum()
    inventory = pd.concat([inventory, data], ignore_index=True)
```

```python
inventory = inventory[["order_placed_at","ingredient_id","qty_available"]]
inventory.rename(columns = {"order_placed_at": "last_update_timestamp"}, inplace=True)
inventory
```
